### Описание данных

В файле `caffeine.csv` содержатся характеристики различных напитков:

* `drink`: название напитка;
* `Volume (ml)`: объем напитка (в миллилитрах);
* `Calories`: количество калорий;
* `Caffeine (mg)`: содержание кофеина (в миллиграммах);
* `type`: тип напитка.
